# `plt.subplots()`

El método `matplotlib.pyplot.subplots` se utiliza para crear una **figura** y un **conjunto de ejes** en los que se pueden trazar gráficos. Este método devuelve dos objetos:

* `fig`: Representa el lienzo o contenedor principalen el que se dibujarán los gráficos.
  
* `ax`: Representa un **conjunto de ejes** dentro de la figura. Los ejes son áreas rectangulares en las que se pueden trazar diferentes tipos de gráficos, como *líneas*, *barras*, *dispersión*, etc. Puedes tener varios conjuntos de ejes en una misma figura si deseas superponer varios gráficos o crear un diseño más complejo.

## `nrows` y `ncols`

### Una figura con una sóla un subgráfica

La función `subplots()` sin argumentos regresa una figura y un único eje. Esta es la forma más sencilla y recomendada de **crear una única figura y ejes**.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Un única subgráfica')
plt.show()
```

### Subgráficas apiladas verticalmente

Los dos primeros argumentos opcionales de `pyplot.subplots` definen el número de filas y columnas de la malla de subgráficas. Cuando se apila en una sola dirección, los ejes devueltos son una matriz numérica 1D que contiene la lista de ejes creados.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, axs = plt.subplots(2)
fig.suptitle('Subgráficas apiladas verticales')
axs[0].plot(x, y)
axs[1].plot(x, -y)
plt.show()
```

**Si está creando solo unos pocos ejes, es útil descomprimirlos inmediatamente en variables dedicadas para cada eje**. De esa manera, podemos usar `ax1` en lugar de `axs[0]`, que es más detallado.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Subgrafica aplilados verticalmente')
ax1.plot(x, y)
ax2.plot(x, -y)
plt.show()
```

### Subgráficas apiladas horizontalmente

Para obtener subgraficas una al lado de la otra, tenemos que definir `nrows=1, ncols=2` para una fila y dos columnas.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
fig.suptitle('Subgraficas apiladas horizontalmente')
ax1.plot(x, y)
ax2.plot(x, -y)
plt.show()
```

### Matriz de Subgráficas

Si queremos crear una **matriz de conjuntos de ejes**, por ejemplo, 2 filas y 2 columnas, basta añadir estos valores como primeros argumentos de la función:

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows = 2, ncols = 5)
plt.show()
```

**Observación:** Los ejes son creados y mostrados, aun cuando no se creen gráficas dentro de ellos.

Vemos en la siguiente imagen que, en este caso, `ax` es un array NumPy, de tamaño 2x2, conteniendo referencias a cada uno de los ejes:

```python
print(type(ax))
print(ax.shape)
print(type(fig))
print(ax)
```

Ahora podríamos ejecutar el método `plot` asociado a cada uno de estos ejes para mostrar una gráfica. Por ejemplo, si quisiéramos mostrarla en la segunda fila (cuyo índice es 1) y primera columna (cuyo índice es 0), podríamos hacerlo del siguiente modo:

```python

import numpy as np


fig, ax = plt.subplots(3,2)
ax[1,0].plot(x,y)
plt.show()
```

### Apilando subplots en dos direcciones

Al apilar en dos direcciones, los ejes devueltos son una matriz 2D NumPy. Si tiene que establecer parámetros para cada subgrafica, es útil iterar sobre todos los subplots en una cuadrícula 2D usando for ax en `axs.flat`

```python

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Eje [0, 0]')
axs[0, 1].plot(x, y, 'orange')
axs[0, 1].set_title('Eje [0, 1]')
axs[1, 0].plot(x, -y, 'green')
axs[1, 0].set_title('Eje [1, 0]')
axs[1, 1].plot(x, -y, 'red')
axs[1, 1].set_title('Eje [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

for ax in axs.flat:
    ax.label_outer()
```

Puede usar el desempaquetado de tuplas también en 2D para asignar todas las subgraficas a variables dedicadas:

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('Sharing x per column, y per row')
ax1.plot(x, y)
ax2.plot(x, y**2, 'orange')
ax3.plot(x, -y, 'green')
ax4.plot(x, -y**2, 'red')

for ax in fig.get_axes():
    ax.label_outer()
```

## `sharex` y `sharey`

Dos parámetros interesantes de la función subplots son sharex y sharey. Éstos controlan la compartición de las propiedades de los ejes. Por defecto toman el valor False, lo que supone que cada conjunto de ejes es independiente. Si, por ejemplo, el argumento sharex se fija a True, todos los ejes x de los diferentes conjuntos de ejes compartirán las mismas propiedades. Veámoslo en la práctica con una matriz de 2x2 conjuntos de ejes:

De forma predeterminada, cada eje se escala individualmente. Por lo tanto, si los rangos son diferentes, los valores de marca de las subgraficas no se alinean

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Los valores de los ejes se escalan individualmente de forma predeterminada')
ax1.plot(x, y)
ax2.plot(x + 1, -y)
plt.show()
```

```python

import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)


fig, ax = plt.subplots(3,2, figsize=(5,5), sharex= True)
ax[1,0].plot(x,y)
plt.show()
```

Comprobamos cómo los ejes horizontales son mostrados solo en la parte inferior de la figura, indicando que los situados por encima de éstos comparten dicho eje.

El equivalente para el eje y es el que podemos ver en la siguiente imagen:

```python

import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)


fig, ax = plt.subplots(2,3, figsize=(5,15), sharey= True)
ax[0,2].plot(x,y)
plt.show()
```

Ahora, los ejes y son mostrados solo en los conjuntos de ejes de la primera columna.

Si fijamos a True ambos parámetros:

```python

import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)


fig, ax = plt.subplots(3,1, figsize=(10,10), sharey= True, sharex= True)
ax[1,0].plot(x,y)
plt.show()
```

Los ejes son mostrados en la primera columna y en la última fila, indicando que el resto de los conjuntos de ejes comparten dichos ejes.

## `Squeeze`

# Método Montecarlo para aproximar $\pi$

El método de Montecarlo es una técnica numérica utilizada para resolver problemas mediante la generación de números aleatorios. Recibe su nombre en honor al famoso casino de Montecarlo en Mónaco, que es conocido por los juegos de azar.

El procedimiento básico del método de Montecarlo implica los siguientes pasos:

* Definir un problema matemático o una pregunta que se desea responder.
  
* Crear un modelo o una simulación del problema. 

* Generar un gran número de muestras aleatorias. Estas muestras representan diferentes escenarios o resultados posibles del problema.

* Aplicar el modelo o la simulación a cada muestra y registrar los resultados.

* Utilizar los resultados de las muestras para estimar o resolver el problema original. Esto puede implicar cálculos estadísticos, promedios, intervalos de confianza u otros métodos de análisis.


La idea clave detrás del método de Montecarlo es que al generar un gran número de muestras aleatorias y aplicar el modelo o la simulación a cada una de ellas, se obtiene una aproximación de la solución o la respuesta deseada. Cuanto mayor sea el número de muestras utilizadas, mayor será la precisión de la estimación.

El método de Montecarlo es especialmente útil en problemas complejos o en aquellos que no se pueden resolver de manera analítica o determinista. Permite abordar problemas que involucran incertidumbre, variabilidad o múltiples variables y proporciona una manera de obtener resultados aproximados cuando no es posible obtener una solución exacta.

Es importante destacar que el método de Montecarlo no garantiza una solución exacta, pero puede proporcionar estimaciones precisas con una suficiente cantidad de muestras aleatorias.

```python
import random
import numpy as np
 

random.seed()

lado = 50 
num_granos = 100

fig, ax = plt.subplots(figsize=(8, 8))
plt.axis("equal")
ax.set_xlim(-lado/2, lado/2)
ax.set_ylim(-lado/2, lado/2)

xs_in, ys_in = list(), list()
xs_out, ys_out = list(), list()


for g in range(num_granos):
        x, y = (np.random.uniform(-1,1))*0.5*lado, (np.random.uniform(-1,1))*lado*0.5
        if x**2 + y**2 <= (lado/2)**2:
            xs_in.append(x)
            ys_in.append(y)
        else:
            xs_out.append(x)
            ys_out.append(y)
            
ax.scatter(xs_in, ys_in, color="r", linewidths=.25)
ax.scatter(xs_out, ys_out, color="b", linewidths=0.25)


circle = plt.Circle((0, 0), lado/2, fill=False, color="g", linewidth=2.5)
ax.add_patch(circle)
ax.set_title("Un experimento con 1000 granos de arena esparcidos", fontsize=15)

plt.show()
```
