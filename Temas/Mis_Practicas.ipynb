```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.array([-3,-1,0,1, 2,3, 4, 5])
y = np.array([1,-5, 3, 10, 8, 1, 5])

# Crear la figura
plt.figure(figsize=(10, 5))

# Ejes coordenados (eje x y eje y en el origen)
plt.axhline(0, color='black', linewidth=1)  # Eje x
plt.axvline(0, color='black', linewidth=1)  # Eje y

# Graficar la función escalonada y líneas verticales punteadas
for i in range(len(x) - 1):
    # Tramo horizontal
    plt.plot([x[i], x[i+1]], [y[i], y[i]], color='blue', linewidth=2, label='Escalón' if i == 0 else "")
    
    # Relleno entre el eje x y el valor de la función
    plt.fill_between([x[i], x[i+1]], [y[i], y[i]], color='lightblue', alpha=0.4)
    
    # Líneas verticales punteadas en extremos del subintervalo
    plt.vlines(x[i], ymin=0, ymax=y[i], linestyle='--', color='gray', linewidth=1)
    plt.vlines(x[i+1], ymin=0, ymax=y[i], linestyle='--', color='gray', linewidth=1)
    
    # Puntos abiertos en los extremos
    plt.plot(x[i], y[i], marker='o', color='blue', markerfacecolor='blue')
    plt.plot(x[i+1], y[i], marker='o', color='blue', markerfacecolor='white')

plt.plot(x[-1], y[-1], marker='o', color='blue', markerfacecolor='blue')


# Configuración del gráfico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Función escalonada')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Mostrar
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import math 

np.random.seed(10)

a = -5; b = 5
k=1000
n=10

def f(x):
    fun = 'np.sin(x)**2'
    return eval(fun)

z = np.linspace(a,b,k)
y = f(z)

plt.axhline(0, linestyle='-', linewidth=1, color='black')
plt.axvline(0, linestyle='-', linewidth=1, color='black')

plt.plot(z,y, color = 'Tomato', linewidth = 1)

x = np.linspace(a,b,n)
sum = 0;
delta = (b-a)/n
k=10
for i  in range(len(x)-1):
    z=np.linspace(x[i],x[i+1],k)
    
    ci = max(np.array([f(w) for w in z]))
    plt.plot([x[i],x[i+1]], [ci,ci], color = 'darkblue', linewidth = 1)
    plt.fill_between([x[i],x[i+1]], [ci,ci], color='blue', alpha=.5)
    plt.vlines([x[i],x[i+1]], ymin = min(0,ci), ymax= max(0,ci), linestyle = '--',color = 'w', linewidth =1 )
    sum += ci*delta 
    
plt.xlabel('x')
plt.ylabel('f(x)=x^2')
plt.title('Suma de Riemann')
plt.show()


print('La suma de Riemann es:',sum)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

# Parámetros
a, b = 0, 5
n = 10

# Función continua en [a, b]
def f(x):
    return math.sqrt(x)

# Generar partición aleatoria ordenada con puntos en (a,b)
internal_points = sorted(random.uniform(a, b) for _ in range(n - 1))
x_partition = [a] + internal_points + [b]

# Generar puntos x_i^* aleatorios dentro de cada subintervalo (x_{i-1}, x_i)
x_stars = [random.uniform(x_partition[i], x_partition[i + 1]) for i in range(n)]
heights = [f(x_star) for x_star in x_stars]

# Crear figura
plt.figure(figsize=(10, 8))
sum = 0
# Graficar los rectángulos estilo escalonado
for i in range(n):
    xi_left, xi_right = x_partition[i], x_partition[i + 1]
    yi = heights[i]
    sum = sum + (xi_right - xi_left)* yi
    # Tramo horizontal azul grueso
    plt.plot([xi_left, xi_right], [yi, yi], color='blue', linewidth=2)

    # Relleno celeste semitransparente
    plt.fill_between([xi_left, xi_right], [0, 0], [yi, yi], color='lightblue', alpha=0.4)

    # Líneas verticales punteadas grises
    plt.vlines([xi_left, xi_right], ymin=0, ymax=yi, linestyle='--', color='gray', linewidth=1)

    # Puntos extremos (círculos blancos con borde azul)
    plt.plot([xi_left, xi_right], [yi, yi], 'o', markerfacecolor='white', markeredgecolor='blue', markersize=4.5)

# Graficar la curva f(x)
x_vals = np.linspace(a, b, 500)
y_vals = [f(x) for x in x_vals]
plt.plot(x_vals, y_vals, color='black')

# Ejes y detalles
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Suma de Riemann de una partición aleatoria')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print('La suma de Riemann es:',sum)

```

# Método Montecarlo para aproximar $\pi$
El método de Montecarlo es una técnica numérica utilizada para resolver problemas mediante la generación de números aleatorios. Recibe su nombre en honor al famoso casino de Montecarlo en Mónaco, que es conocido por los juegos de azar.
El procedimiento básico del método de Montecarlo implica los siguientes pasos:

* Definir un problema matemático o una pregunta que se desea responder.
  
* Crear un modelo o una simulación del problema. 

* Generar un gran número de muestras aleatorias. Estas muestras representan diferentes escenarios o resultados posibles del problema.

* Aplicar el modelo o la simulación a cada muestra y registrar los resultados.

* Utilizar los resultados de las muestras para estimar o resolver el problema original. Esto puede implicar cálculos estadísticos, promedios, intervalos de confianza u otros métodos de análisis.


La idea clave detrás del método de Montecarlo es que al generar un gran número de muestras aleatorias y aplicar el modelo o la simulación a cada una de ellas, se obtiene una aproximación de la solución o la respuesta deseada. Cuanto mayor sea el número de muestras utilizadas, mayor será la precisión de la estimación.

El método de Montecarlo es especialmente útil en problemas complejos o en aquellos que no se pueden resolver de manera analítica o determinista. Permite abordar problemas que involucran incertidumbre, variabilidad o múltiples variables y proporciona una manera de obtener resultados aproximados cuando no es posible obtener una solución exacta.

Es importante destacar que el método de Montecarlo no garantiza una solución exacta, pero puede proporcionar estimaciones precisas con una suficiente cantidad de muestras aleatorias.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

random.seed()

lado = 50 
num_granos = 100

plt.figure(figsize=(8, 8))
plt.axis("equal")
plt.xlim(-lado/2, lado/2)
plt.ylim(-lado/2, lado/2)

xs_in, ys_in = [], []
xs_out, ys_out = [], []

for g in range(num_granos):
    x = np.random.uniform(-1, 1) * 0.5 * lado
    y = np.random.uniform(-1, 1) * 0.5 * lado
    if x**2 + y**2 <= (lado/2)**2:
        xs_in.append(x)
        ys_in.append(y)
    else:
        xs_out.append(x)
        ys_out.append(y)

# Usamos plot para simular scatter
plt.scatter(xs_in, ys_in, color='red', s=2)  # 'r' = red, 'o' = marker circular
plt.scatter(xs_out, ys_out, color='b', s=2) # 'b' = blue

# Dibujo del círculo usando puntos
theta = np.linspace(0, 2 * np.pi, 300)
x_circle = (lado / 2) * np.cos(theta)
y_circle = (lado / 2) * np.sin(theta)
plt.plot(x_circle, y_circle, 'g-', linewidth=2.5)

plt.title("Un experimento con 100 granos de arena esparcidos", fontsize=15)
plt.show()
```
