# "La caminata del Actuario en Huatulco"

Imaginemos por un momento que estamos saliendo de una noche de fiesta intensa en Papaya, ese antro famoso de Huatulco que muchos conocen. Todo ha sido música, luces, y buena compañía… pero ahora, nuestro protagonista de esta historia —llamémosle “El Actuario”— decide volver caminando a la playa de Chahué.

Pero hay un detalle importante: el actuario no tiene rumbo fijo. Su sentido de orientación se ha desvanecido entre el desvelo y el baile, y cada paso que da es completamente aleatorio. Un paso hacia adelante, luego tal vez a la izquierda, después quizás hacia atrás… nadie puede predecir con certeza qué dirección tomará.

Esta caminata caótica pero fascinante es lo que en matemáticas conocemos como una ***caminata aleatoria***. *Cada paso es independiente del anterior*, y no hay una meta predefinida.

**¿Llegará finalmente a la playa? ¿Terminará de vuelta en Papaya? ¿O se perderá por las calles de Huatulco?**


```python
import numpy as np
import matplotlib.pyplot as plt

# Número de pasos
n = 100

# Generar pasos aleatorios: +1 o -1 con igual probabilidad
pasos = np.random.choice([-1, 1], size=n)

# Posiciones acumuladas (inicio en 0)
posicion = np.cumsum(pasos)
posicion = np.insert(posicion, 0, 0)  # Para empezar desde el origen

# Eje de tiempo (pasos)
tiempo = np.arange(n + 1)

# Graficar la caminata
plt.figure(figsize=(10, 5))
plt.plot(tiempo, posicion, marker='o', linestyle='-', color='blue', label='Caminata Aleatoria')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.xlabel('Paso')
plt.ylabel('Posición')
plt.title('Caminata Aleatoria en 1D')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la simulación
num_caminatas = 10     # Número de trayectorias
num_pasos = 1000         # Número de pasos por caminata
objetivo=100

# Crear figura
plt.figure(figsize=(12, 6))
llego_a_la_playa = 0
for _ in range(num_caminatas):
    pasos = np.random.choice([-1, 1], size=num_pasos, p=[0.5, 0.5])  # Pasos aleatorios
    posicion = np.cumsum(pasos)                        # Posición acumulada
    posicion = np.insert(posicion, 0, 0)               # Inicia en la posición 0
    plt.plot(posicion, alpha=0.6, linewidth=1)
    
    if np.any(posicion >= objetivo):  # ¿En algún momento llegó o pasó?
        llego_a_la_playa += 1

# Estimar probabilidad
probabilidad = llego_a_la_playa / num_caminatas

print(f"Probabilidad estimada de que el Actuario llegue a la playa de Chahué: {probabilidad:.3f}")


# Añadir elementos descriptivos
plt.title("La caminata del Actuario en Huatulco", fontsize=14)
plt.xlabel("Número de pasos")
plt.ylabel("Posición")
plt.axhline(0, color='black', linestyle='--', linewidth=1)  # Línea base
plt.axhline(objetivo, color='cyan', linestyle='--', linewidth=1)  # Línea base
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()

# Mostrar gráfico
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la simulación
num_caminatas = 1000    # Número de trayectorias
num_pasos = 1000        # Número de pasos por caminata
objetivo = 100          # Posición objetivo: la playa de Chahué

# Contadores
llego_a_la_playa = 0
regreso_a_papaya = 0

# Crear figura
plt.figure(figsize=(12, 6))

for _ in range(num_caminatas):
    pasos = np.random.choice([-1, 1], size=num_pasos, p=[0.5, 0.5])  # Pasos aleatorios
    posicion = np.cumsum(pasos)              # Posición acumulada
    posicion = np.insert(posicion, 0, 0)     # Inicia en la posición 0

    
    plt.plot(posicion, alpha=0.6, linewidth=1)

    # ¿Llegó o pasó por la playa?
    if np.any(posicion >= objetivo):
        llego_a_la_playa += 1

    # ¿Regresó a Papaya (posición 0) en algún momento después del inicio?
    if np.any(posicion[1:] == 0):  # Excluimos el primer paso que es siempre cero
        regreso_a_papaya += 1

# Estimar probabilidades
prob_playa = llego_a_la_playa / num_caminatas
prob_papaya = regreso_a_papaya / num_caminatas

print(f"Probabilidad estimada de que el Actuario llegue a la playa de Chahué (posición {objetivo}): {prob_playa:.3f}")
print(f"Probabilidad estimada de que el Actuario regrese a Papaya (posición 0): {prob_papaya:.3f}")

# Añadir elementos descriptivos
plt.title("La caminata del Actuario en Huatulco", fontsize=14)
plt.xlabel("Número de pasos")
plt.ylabel("Posición")
plt.axhline(0, color='black', linestyle='--', linewidth=1)  # Línea base
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()

# Mostrar gráfico (solo si graficaste caminatas)
plt.show()
```
