```python
columnas=["IMDb","Rotten Tomatoes","Directors"]
def valores(indice):
    if indice==columnas[0]:
        valor="Sin calificar"
    elif indice==columnas[1]:
        valor="Sin clasificación"
    elif indice==columnas[2]:
        valor=" "
    else:
        valor=" "
    return valor
    
```
