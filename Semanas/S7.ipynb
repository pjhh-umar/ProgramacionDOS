# Librería `Matplotlib`



**Matplotlib** es una librería de Python especializada en la creación de gráficos en dos dimensiones.

Se trata de un paquete grande y relativamente compleja que contiene dos módulos principales: `pyplot` y `pylab`

`Pyplot` ofrece una interfaz fácil para crear gráficos fácilmente, automatizando la creación de figuras y ejes cuando hace se crea un gráfico. Por otra parte, `pylab` combina pyplot para crear gráficos con funcionalidad de `numpy` y hacer cálculos con arrays usando un único `espacio de nombres` muy parecido a `Matlab`.

```python
import matplotlib.pyplot as plt

ahorros = [50, 100, 30, 65, 120,0]

plt.plot(ahorros)
plt.show()
```

```python
import math
x = list(range(-10,10))
y = [ x**3 for x in x]
plt.plot(x, y,color='blue', linewidth=1.5, label='$f(x)=x^{3}$')
plt.legend()
plt.title('Mi gráfica')
plt.ylabel('Eje Y') #Etiqueta para eje Y
plt.xlabel('Eje X') #Etiqueta para eje X
plt.grid()
plt.show()

```

```python
from matplotlib import pyplot

# Función cuadrática.
def f1(x):
    return 2*(x**2) + 5*x - 2

# Función lineal.
def f2(x):
    return 4*x + 1

# Valores del eje X que se toman el gráfico.
x = range(-10, 10)

# Graficar ambas funciones.
pyplot.plot(x, [f1(i) for i in x])
pyplot.plot(x, [f2(i) for i in x])

# Limitar los valores de los ejes.
pyplot.xlim(-10, 10)
pyplot.ylim(-10, 10)

# Establecer el color de los ejes.
pyplot.axhline(0,color="black")
pyplot.axvline(0, color="black")

# Guardar gráfico como imágen PNG.
pyplot.savefig("mi_grafica.ps")
pyplot.show()
```

## Diagrama de línea

```python
import matplotlib.pyplot as plt

#Definir los datos
x1 = [3, 4, 5, 6]
y1 = [5, 6, 3, 4]
x2 = [2, 5, 8]
y2 = [3, 4, 3]

#Configurando las características del gráfico
plt.plot(x1, y1, label = 'Línea 1', linewidth = 1, color = 'blue', linestyle="--", marker="*", markersize="5", alpha = 1)
plt.plot(x2, y2, label = 'Línea 2', linewidth = 1, color = 'green', linestyle = ":")

#Definir título y nombres de ejes
plt.title('Diagrama de Líneas')
plt.ylabel('Eje Y')
plt.xlabel('Eje X')

#Mostrar leyenda, cuadrícula y figura
plt.legend()
plt.grid()
plt.show()
```

## Gráfica de barras

```python
#Definir los datos
x1 = ['A','B','C', 'D', 'E']
y1 = [10, 55, 80, 32, 40]
x2 = ['a', 'b', 'c', 'd', 'e']
y2 = [42, 26, 10, 29, 66]

#Configurar las características del gráfico
plt.bar(x1, y1, label = 'Datos 1', width = 0.75, color = 'blue')
plt.bar(x2, y2, label = 'Datos 2', width = 0.75, color = 'orange')

#Definir título y nombres de ejes
plt.title('Gráfico de barras')
plt.ylabel('Valores')
plt.xlabel('Categorias')

#Mostrar leyenda y figura
plt.legend()
plt.show()
```

```python
#Definir los datos
x1 = [0.25, 1.25, 2.25, 3.25, 4.25]
y1 = [10, 55, 80, 32, 40]
x2 = [0.75, 1.75, 2.75, 3.75, 4.75]
y2 = [42, 26, 10, 29, 66]

#Configurar las características del gráfico
plt.bar(x1, y1, label = 'Datos 1', width = 0.25, color = 'blue')
plt.bar(x2, y2, label = 'Datos 2', width = 0.75, color = 'orange')

#Definir título y nombres de ejes
plt.title('Gráfico de barras')
plt.ylabel('Categorias')
plt.xlabel('Valores')

#Mostrar leyenda y figura
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 25]

# Cambiar el color y el ancho de las barras
colores = ['red', 'green', 'blue', 'orange'] #Lista de colores, para asignar un color diferente a cada barra.
anchos = [0.5, 0.5, 0.5, 0.5] #Se establece el ancho de cada barra

# Crear el gráfico de barras con ajustes
plt.bar(categorias, valores, color=colores, width=anchos)

# Personalizar el gráfico
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfica de barras')

# Mostrar el gráfico
plt.show()
```

## Gráfico circular

```python
#Definir los datos
dormir =[8,7,8,7.5,7.5]
comer = [1.5,1.5,1.5,1.5,1.5]
trabajar =[11,11,9.5,8,11]
recreacion = [1,0,2,2,0]
total=sum(dormir) + sum(comer) + sum(trabajar) + sum(recreacion)

divisiones = [sum(dormir)/total,sum(comer)/total,sum(trabajar)/total, sum(recreacion)/total]
actividades = ['Dormir','Comer','Trabajar','Recreación']
colores = ['red','purple','blue','orange']

#Configurar las características del gráfico
plt.pie(divisiones, labels=actividades, colors=colores, startangle=45, shadow=True, explode=(0.0,0.25,0.0,0), autopct='%1.2f%%')

#Definir título
plt.title('Estilo de vida de los estudiantes del 2do semestre de Actuaría en UMAR :D')

#Mostrar figura
plt.show()
```
