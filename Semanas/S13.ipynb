# Dataframe

Un objeto del tipo `DataFrame` define un conjunto de datos estructurado en forma de tabla donde cada columna es un objeto de tipo `Series`. En general, todos los datos de una misma columna son del mismo tipo, y las filas son registros que pueden contender datos de distintos tipos.

Un `DataFrame` contiene dos índices, uno para las *filas* y otro para las *columnas*, y se puede acceder a sus elementos mediante los nombres de las filas y las columnas.

## Creación de un DataFrame

````python
import pandas as pd
````

````python
dataframe = pd.DataFrame({'A': [1, 2, 3]})
pd.DataFrame({'A': [1, 2, 3]})
print(dataframe)
````

### Creación de un DataFrame a partir de un diccionario de listas

Para crear un `DataFrame` a partir de un diccionario cuyas claves son los nombres de las columnas y los valores son listas con los datos de las columnas se utiliza el método:

```python
pd.Dataframe(data=diccionario, index=filas, columns=columnas, dtype=tipos)
```

* Devuelve un objeto del tipo DataFrame cuyas columnas son las listas contenidas en los valores del diccionario `diccionario`,
  
* Los nombres de filas son los indicados en la lista `filas`,

* Los nombres de columnas son los indicados en la lista `columnas` y los tipos indicados en la lista tipos.

* La lista `filas` tiene que tener el mismo tamaño que las listas del diccionario, mientras que las listas `columnas` y tipos tienen que tener el mismo tamaño que el diccionario.
  
* Si no se pasa la lista de filas se utilizan como nombres los enteros empezando en 0. Si no se pasa la lista de columnas se utilizan como nombres las claves del diccionario. Si no se pasa la lista de tipos, se infiere.

**Observación:** Los valores asociados a las claves del diccionario deben ser listas del mismo tamaño.

````python
import pandas as pd
datos = {'Nombre':['María', 'Luis', 'Carmen', 'Antonio'],
         'Edad':[18, 22, 20, 21],
         'Carrera':['Economía', 'Medicina', 'Arquitectura', 'Economía'],
         'Correo':['maria@gmail.com', 'luis@yahoo.es', 'carmen@gmail.com', 'antonio@gmail.com']
         }
df = pd.DataFrame(datos)
df
#print(df)
````

````python
# @title Edad

from matplotlib import pyplot as plt
df['Edad'].plot(kind='hist', bins=20, title='Edad')
plt.gca().spines[['top', 'right',]].set_visible(False)
````

### Creación de un DataFrame a partir de una listas de listas

Para crear un `DataFrame` a **partir de una lista de listas** con los datos de las columnas se utiliza el siguiente método:

`pd.Dataframe(data=listas, index=filas, columns=columnas, dtype=tipos)`

* Devuelve un objeto del tipo `DataFrame` cuyas columnas son los valores de las listas de la lista listas,

* Los nombres de filas indicados en la lista `filas`, los nombres de columnas indicados en la lista `columnas` y los tipos indicados en la lista tipos.
  
* La lista `filas`, tiene que tener el mismo tamaño que la lista listas mientras que las listas `columnas` y tipos tienen que tener el mismo tamaño que las listas anidadas en `listas`.
  
* Si no se pasa la lista de `filas` o de `columnas` se utilizan enteros empezando en 0. Si no se pasa la lista de tipos, se infiere.

**Observación:** Si las listas anidadas en listas no tienen el mismo tamaño, las listas menores se rellenan con valores `NaN`.

````python
import pandas as pd
df = pd.DataFrame([['María', 18, 'Economia'], ['Luis', 22, 'Actuaria'], ['Carmen', 20]], columns=['Nombre', 'Edad', 'Carrera'])
print(df)
````

### Creación de un DataFrame a partir de una lista de diccionarios

Para crear un DataFrame a partir de una lista de diccionarios con los datos de las filas, se utiliza el siguiente método:

`pd.Dataframe(data=diccionarios, index=filas, columns=columnas, dtype=tipos)`

* Devuelve un objeto del tipo DataFrame cuyas filas contienen los valores de los diccionarios de la lista diccionarios,

* Los nombres de filas indicados en la lista filas, los nombres de columnas indicados en la lista columnas y los tipos indicados en la lista tipos.

* La lista filas tiene que tener el mismo tamaño que la lista lista. Si no se pasa la lista de filas se utilizan enteros empezando en 0.
  
* Si no se pasa la lista de columnas se utilizan las claves de los diccionarios.

**Observación:** Si no se pasa la lista de tipos, se infiere. Si los diccionarios no tienen las mismas claves, las claves que no aparecen en el diccionario se rellenan con valores `NaN`.

````python
import pandas as pd
df = pd.DataFrame([{'Nombre':'María', 'Edad':18}, {'Edad':22, 'Nombre':'Luis', }, {'Nombre':'Carmen'}])
print(df)
````

## Renombrar los nombres de las filas y columnas

Para cambiar el nombre de las filas y las columnas de un DataFrame se utiliza el siguiente método:

```python
DataFrame.rename(mapper=None, *, index=None, columns=None, axis=None, copy=True, inplace=False, level=None, errors='ignore')
```

donde

* **mapper**: Diccionario o función para aplicar a las etiquetas.

* **index**: Diccionario o función para renombrar las etiquetas de filas.

* **columns**: Diccionario o función para renombrar las etiquetas de columnas.

* **axis**: Especifica el eje para aplicar el mapeo (0 o 'index' para filas, 1 o 'columns' para columnas).

* **copy**: Si es True, se copia el DataFrame original antes de hacer las modificaciones.

* **inplace**: Si es True, hace las modificaciones en el DataFrame original sin devolver uno nuevo.

* **level**: Si el eje es un MultiIndex (índice múltiple), especifica el nivel a aplicar.

* **errors**: Controla el manejo de errores en caso de etiquetas no encontradas. Las opciones son 'ignore' y 'raise'.

Devuelve el DataFrame que resulta de renombrar las columnas indicadas en las claves del diccionario columnas con sus valores y las filas indicadas en las claves del diccionario filas con sus valores en el DataFrame df.

````python
import pandas as pd
df = pd.read_csv('/content/colesterol.csv', sep = ';',decimal=',')
nuevodf =  df.rename(columns={'nombre':'Nombre y apellidos', 'altura':'estatura'}, index={0:1000, 1:1001, 2:1002})
print(nuevodf)
````

### Renombrar Columnas

Para renombrar columnas, puedes usar el parámetro columns con un diccionario que mapea los nombres actuales a los nuevos nombres.

````python
import pandas as pd

# Crear un DataFrame de ejemplo
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

print("DataFrame Original:")
print(df)

# Renombrar columnas usando el método rename
df_renamed = df.rename(columns={'A': 'Alpha', 'B': 'Bravo', 'C': 'Charlie'})
print("\nDataFrame con Columnas Renombradas:")
print(df_renamed)
````

### Renombrar Filas

Para renombrar las etiquetas de las filas, usa el parámetro index.

````python
# Crear un DataFrame con índices específicos
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}, index=['row1', 'row2', 'row3'])

print("\nDataFrame Original:")
print(df)

# Renombrar filas usando el método rename
df_renamed = df.rename(index={'row1': 'first', 'row2': 'second', 'row3': 'third'})
print("\nDataFrame con Filas Renombradas:")
print(df_renamed)
````

### Uso de `inplace`

Usar `inplace` para Modificar el DataFrame Original. El parámetro `inplace` te permite hacer los cambios directamente en el DataFrame original sin crear una copia.

````python
# Renombrar columnas en el DataFrame original
df.rename(columns={'A': 'Alpha', 'B': 'Bravo'}, inplace=True)
print("\nDataFrame Original Modificado:")
print(df)
````

### Renombrar Usando Funciones

Puedes usar funciones para aplicar transformaciones a las etiquetas de las columnas o filas.

````python
# Crear un DataFrame de ejemplo
df = pd.DataFrame({
    'column1': [1, 2, 3],
    'column2': [4, 5, 6]
})

print("\nDataFrame Original:")
print(df)

# Renombrar columnas usando una función
df_renamed = df.rename(columns=lambda x: x.upper())
print("\nDataFrame con Columnas Renombradas a Mayúsculas:")
print(df_renamed)
````

## Cambiar el índice de un DataFrame

Aunque el índice de un DataFrame suele fijarse en la creación del mismo, en ocasiones puede ser necesario cambiar el índice una vez creado el DataFrame. Para ello se utiliza el siguiente método:

<center><code>df.set_index(keys = columnas, verify_integrity = bool, drop=True, inplace=False)</code></center>

* Devuelve el `DataFrame` que resulta de eliminar las columnas de la lista columnas y convertirlas en el nuevo índice.
  
* El parámetro `verify_integrity` recibe un booleano (`False` por defecto) y realiza una comprobación para evitar duplicados en la clave cuando recibe `True`.

* `drop`: Es un valor booleano que indica si las columnas seleccionadas deben eliminarse del DataFrame. Si se establece en `True`, las columnas se eliminarán; si se establece en `False`, las columnas seleccionadas permanecerán en el DataFrame.

* `inplace`: Es un valor booleano que indica si se debe modificar el DataFrame original. Si se establece en `True`, el DataFrame original se modificará y no se devolverá un nuevo DataFrame; si se establece en `False`, se devolverá un nuevo DataFrame con el índice modificado y el DataFrame original no se verá afectado.

### Método `pandas.to_datatime`

```python
pd.to_datetime(arg, errors='raise', dayfirst=False, yearfirst=False, utc=False, format=None, exact=True, unit=None, infer_datetime_format=False, origin='unix', cache=True)

```

* **arg**: La entrada a convertir en datetime. Puede ser una serie de pandas, un array, una lista, o un DataFrame con columnas de año, mes y día

* **format**: Una cadena de formato para especificar el formato de las fechas (por ejemplo, %d/%m/%Y).

* **errors**: Cómo manejar errores. Se tienen las siguientes opciones:
  
  * `raise`: Lanza una excepción en caso de error (comportamiento predeterminado).
  
  * `coerce`: Convierte valores no válidos en `NaT`.
  
  * `ignore`: Retorna los datos sin modificar.

* **dayfirst**: Booleano. Si `True`, se considera que el primer valor es el día (en formatos como DD/MM/YYYY)

* **yearfirst**: Booleano. Si `True`, se considera que el primer valor es el año (en formatos como YYYY/DD/MM).

#### Convertir una Serie de Fechas en String a datetime

````python
dates = pd.Series(['2024-06-01', '2024-06-02', '2024-06-03'])
dates
dates_datetime = pd.to_datetime(dates)
dates_datetime
````

#### Manejar Diferentes Formatos de Fecha

````python
dates = pd.Series(['2024-06-01', '2024-06-02', '2024-06-03'])

dates_datetime = pd.to_datetime(dates, format='%d/%m/%Y')
print(dates_datetime)
````

### Fusionar datos de Dataframes distintos: `pandas.concat()`

El método `concat` en pandas se utiliza para concatenar pandas `objects` como Series y DataFrames a lo largo de un eje (axis).

La sintaxis general es:

```python
pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)
```

* **objs**: Una lista o diccionario de objetos pandas (Series o DataFrame) que se desea concatenar.

* **axis**: Eje a lo largo del cual concatenar. `0` para concatenar por filas, `1` para concatenar por columnas. El valor predeterminado es `0`.

* **join**: Tipo de unión. Puede ser `inner` o `outer`. El valor predeterminado es `outer`.

* **ignore_index**: Si se debe ignorar el índice del eje de concatenación. El valor predeterminado es `False`.
  
* **keys**: Secuencia para usar como jerarquía de los niveles en el índice del eje resultante.

* **sort**: Ordenar columnas si el eje es `1`. No es necesario si las columnas ya están ordenadas. El valor predeterminado es `False`.

### Concatenar DataFrames por Filas (`axis=0`):


````python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']})
df2 = pd.DataFrame({'A': ['A3', 'A4', 'A5'], 'B': ['B3', 'B4', 'B5']})

print(df1)
print(df2)
````

````python
resultado = pd.concat([df1, df2])
print(resultado)
````

### Concatenar DataFrames por Columnas (`axis=1`):

````python
result = pd.concat([df1, df2], axis=1)
print(result)
````

### Concatenar y Ignorar el Índice:

````python
result = pd.concat([df1, df2], ignore_index=True)
print(result)
````

### Concatenar con Unión Interna (`inner join`):

````python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']})
df2 = pd.DataFrame({'B': ['B3', 'B4', 'B5'], 'C': ['C3', 'C4', 'C5']})

result = pd.concat([df1, df2])
print(result)
````

````python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2'], 'Z': ['Z0', 'Z1', 'Z2']})
df2 = pd.DataFrame({'B': ['B3', 'B4', 'B5'], 'A': ['C3', 'C4', 'C5']})

result = pd.concat([df1, df2], join='inner')
print(result)
````

````python
import pandas as pd
resultado = pd.concat([df1, df2], keys=['Dataframe1', 'Dataframe2'])
print(resultado)
````

````python
import pandas as pd
resultado = pd.concat([df1, df2], verify_integrity=True)
print(resultado)
````

La sintaxis general del método `data_range` es:

```python
pd.date_range(start=None, end=None, periods=None, freq='D', tz=None, normalize=False, name=None, closed=None, **kwargs)
```
donde
* **start**: La fecha de inicio del rango. Puede ser una cadena, `datetime`, o `Timestamp`.

* **end**: La fecha de finalización del rango. Similarmente, puede ser una `cadena`, `datetime`, o `Timestamp`.

* **periods**: Número de periodos a generar. Exclusivo con los parámetros `start` y `end`.

* **freq**: Frecuencia de los periodos. Los más comunes son:

  * **'D'**: Día (por defecto).

  * **'B'**: Día hábil.

  * **'H'**: Hora.

  * **'T'** o **'min'**: Minuto.

  * **'S'**: Segundo.

  * **'A'** o **'Y'**: Año fin de período.

  * **'W'**: Semana.

  * **'M'**: Mes fin de período

<!-- * **'AS'** o **'YS'**: Año inicio de período. -->

<!-- * **'Q'**: Trimestre fin de período. -->

<!-- * **'QS'**: Trimestre inicio de período. -->


* **name**: Nombre de la serie resultante.

## `pandas.Timestamp`

El método `Timestamp` en pandas es una clase que **representa un sólo punto en el tiempo**, similar a un objeto datetime en Python pero con mayor funcionalidad y optimización para su uso en análisis de datos.

### Creación de un `Timestamp`

Puedes crear un `Timestamp` de varias maneras:

#### Desde una cadena de fecha y hora:

````python
import pandas as pd

ts = pd.Timestamp('2023-05-27')
print(ts)
````

#### Desde componentes individuales (año, mes, día, hora, minuto, segundo.):

````python
ts = pd.Timestamp(2023, 5, 27, 10, 30, 45)
print(ts)
````

#### Desde un objeto datetime de Python:

````python
from datetime import datetime

dt = datetime(2023, 5, 27, 10, 30, 45)
print('dt = ', dt)
ts = pd.Timestamp(dt)
print(ts)
````

## Propiedades y Métodos de Timestamp

### Propiedades

````python
ts = pd.Timestamp('2023-05-27 10:30:45')
print("Año:", ts.year)
print("Mes:", ts.month)
print("Día:", ts.day)
print("Hora:", ts.hour)
print("Minuto:", ts.minute)
print("Segundo:", ts.second)
print("Día de la semana:", ts.dayofweek)  # Lunes=0, Domingo=6
print("Nombre del día de la semana:", ts.day_name())  # Nombre del día en inglés :(
````

### Algunos **métodos**

Ajustar el tiempo a la medianoche

````python
ts_midnight = ts.normalize()
print("Medianoche:", ts_midnight)
````

#### Convertir a otra zona horaria

````python
ts_utc = ts.tz_localize('UTC')
print("Zona horaria UTC:", ts_utc)
````

#### Convertir a una zona horaria diferente


````python
ts_pacific = ts_utc.tz_convert('US/Pacific')
print("Zona horaria Pacífico:", ts_pacific)
````

## Operaciones con Timestamp

### Comparación de fechas y horas:

````python
ts1 = pd.Timestamp('2023-05-27 10:30:45')
ts2 = pd.Timestamp('2023-05-28 10:30:45')

print("ts1 es antes de ts2:", ts1 < ts2)
print("ts1 es igual a ts2:", ts1 == ts2)
print("ts1 es después de ts2:", ts1 > ts2)
````

## `pandas.Categorical`

El método `Categorical` en pandas se utiliza para trabajar con datos categóricos, que son datos que pueden tomar un número limitado y generalmente fijo de posibles valores (categorías). Los datos categóricos se pueden utilizar para realizar operaciones más eficientes y ahorrar memoria, ya que en lugar de almacenar cada valor repetidamente, solo se almacenan referencias a las categorías.

## Creación de Datos Categóricos

Puedes crear datos categóricos de varias maneras:

### Convertir una Serie existente a datos categóricos:

````python
import pandas as pd

serie = pd.Series(['manzana', 'pera', 'manzana', 'naranja', 'pera'])
serie_categorica = serie.astype('category')
print(serie_categorica)
````

### Crear una Serie categórica directamente:

````python
import pandas as pd

serie_categorica = pd.Categorical(['manzana', 'pera', 'manzana', 'naranja', 'pera'])
print(serie_categorica)
````

### Crear una Serie categórica con categorías específicas y ordenadas:

````python
import pandas as pd

serie_categorica = pd.Categorical(['manzana', 'pera', 'manzana', 'naranja', 'pera'],
                               categories=['naranja', 'pera', 'manzana'],
                               ordered=True)
print(serie_categorica)
````

## Propiedades y Métodos de los Datos Categóricos

Los datos categóricos tienen varias propiedades y métodos útiles:

* **categories**: Devuelve las categorías.
  
* **codes**: Devuelve los códigos de las categorías.
  
* **ordered**: Indica si las categorías tienen un orden.

````python
print("Categorías:", serie_categorica.categories)
print("Códigos:", serie_categorica.codes)
print("Ordenado:", serie_categorica.ordered)
````

## Operaciones con Datos Categóricos

### Comparación:

````python
print(serie_categorica > 'pera')
````

### Agrupación:

````python
df = pd.DataFrame({
    'frutas': ['manzana', 'pera', 'manzana', 'naranja', 'pera'],
    'cantidad': [10, 5, 8, 12, 7]
})
df['frutas'] = df['frutas'].astype('category')
grouped = df.groupby('frutas').sum()
print(grouped)
````

### Conteo de valores:

````python
print(serie_categorica.value_counts())
````

## `Dataframe.groupby `

El método `groupby` en pandas se utiliza para agrupar datos en un DataFrame según uno o más criterios. Esto permite realizar operaciones de agregación, transformación y filtrado en grupos de datos.

```python
DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=False, observed=False, dropna=True)
```

donde

* **by**: Criterios para agrupar los datos. Puede ser una columna, una lista de columnas, una función, un dict, una Series, etc.
  
* **axis**: Eje a lo largo del cual se agrupan los datos. El valor predeterminado es 0 (filas).

* **level**: Si el índice es jerárquico (MultiIndex), agrupa por uno o más niveles.

* **as_index**: Si es True, los valores agrupados se utilizan como el índice del resultado. El valor predeterminado es True.

* **sort**: Si es True, los grupos se ordenan por sus etiquetas. El valor predeterminado es True.

* **group_keys**: Si es True, los keys de grupo se incluyen en el índice del resultado. El valor predeterminado es True.


### Agrupar por una Columna y Calcular la Media

````python
import pandas as pd

data = {
    'Equipo': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Puntos': [10, 20, 15, 25, 10, 30]
}
df = pd.DataFrame(data)

group = df.groupby('Equipo')
print(group.mean())
````

### Agrupar por Múltiples Columnas

````python
data = {
    'Equipo': ['A', 'A', 'B', 'B', 'A', 'A'],
    'Jugador': ['K1', 'K2', 'K1', 'K2', 'K1', 'K2'],
    'Puntos': [5, 10, 15, 20, 25, 30]
}
df = pd.DataFrame(data)

group = df.groupby(['Equipo', 'Jugador'])
print(group.sum())
````

### Aplicar Funciones de Agregación Personalizadas

````python
def rango(x):
    return x.max() - x.min()

group = df.groupby('Equipo')
print(group.agg(rango))
````

### Filtrar Grupos

````python
def filtro(x):
    return x['Puntos'].mean() > 20

group = df.groupby('Equipo').filter(filtro)
print(group)
````

### Transformar Grupos

````python
df['Media_Puntos'] = df.groupby('Equipo')['Puntos'].transform('mean')
print(df)
````

### Aplicar Funciones a Grupos y Expandir el Resultado

````python
def sumar_puntos(grupo):
    return grupo.assign(Total=grupo['Puntos'].sum())

group = df.groupby('Equipo').apply(sumar_puntos)
print(group)
````
